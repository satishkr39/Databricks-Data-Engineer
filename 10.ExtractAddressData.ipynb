# Read the address files which is TSV format
- We will use read_files function to overcome the limitation of noraml reading files
- Create address view in bronze layer

```sql
-- limitations of using normal CSV read file
select * FROM csv.`/Volumes/gizmbobox/landing/landing/operational_data/addresses/`
```

Link for documentation: https://learn.microsoft.com/en-us/azure/databricks/sql/language-manual/functions/read_files

```sql
 SELECT * FROM read_files('/Volumes/gizmbobox/landing/landing/operational_data/addresses/',
                        format => 'csv',
                        delimiter => '\t',
                        header => true)
```

- Create schema

```sql
CREATE OR REPLACE VIEW gizmbobox.bronze.v_address AS 
SELECT * FROM read_files('/Volumes/gizmbobox/landing/landing/operational_data/addresses/',
                        format => 'csv',
                        delimiter => '\t',
                        header => true)
```
