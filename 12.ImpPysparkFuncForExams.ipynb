
## Important for Databricks Exam
- Run SQL Command using SQL function
- Spark dataframe reader API
- Read tables suing Spark.table function

```python
%python 

# Run sql command
df = spark.sql('SELECT * FROM json.`/Volumes/gizmbobox/landing/landing/operational_data/customers/`')
df.display()
```

```python
# dataframe reader API
df_1 = spark.read.format('json').load('/Volumes/gizmbobox/landing/landing/operational_data/customers/')
df_1.display()
```

```python
# read the tabel directly

kdf_table = spark.table('gizmbobox.bronze.payments')
kdf_table.display()
```
