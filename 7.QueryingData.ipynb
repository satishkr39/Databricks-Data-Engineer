## Creating Views

```sql
USE CATALOG gizmbobox;
USE SCHEMA bronze;
```

```sql

-- Create View Now 
CREATE OR REPLACE VIEW gizmbobox.bronze.V_customers AS
SELECT *, _metadata.file_path FROM JSON.`dbfs:/Volumes/gizmbobox/landing/landing/operational_data/customers/`
```

```sql
-- Query the view
select * from v_customers;
```
