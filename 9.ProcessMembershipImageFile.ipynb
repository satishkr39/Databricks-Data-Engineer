# Reading membership folders
- Query and process the image as binary file

```sql
%fs ls '/Volumes/gizmbobox/landing/landing/operational_data/memberships/'
```

- Reading the binary file format

```sql
-- */* means wildcard characters. To fetch all the images within the subdirectory. 
SELECT * FROM binaryFile.`/Volumes/gizmbobox/landing/landing/operational_data/memberships/*/*.png`
```

 - Create the membership view

```sql
CREATE OR REPLACE VIEW gizmbobox.bronze.v_membership AS 
SELECT * FROM binaryFile.`/Volumes/gizmbobox/landing/landing/operational_data/memberships/*/*.png`
```
