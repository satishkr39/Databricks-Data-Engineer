# Extract Orders File
- Query order files using JSON
- Query order files using text
- Create order view in bronze schema 

```sql
-- Read order files
-- We will get a new column named corrupt_recrod when our files are having issue with syntax. 
select * from json.`/Volumes/gizmbobox/landing/landing/operational_data/orders/`
```

```sql
-- to fix the above corrupt_data issue: we will read the file as text and pre-process data and fix the problem. 
-- Read the json file as text now
SELECT * FROM text.`/Volumes/gizmbobox/landing/landing/operational_data/orders/`
```

```sql
-- Create order view in bronze schema
CREATE OR REPLACE VIEW gizmbobox.bronze.v_orders AS
SELECT * FROM text.`/Volumes/gizmbobox/landing/landing/operational_data/orders/`
```
