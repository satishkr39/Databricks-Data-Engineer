# Creating External Table
- List the files from payment folder
- Create external table
- Demo effect of adding/updating/delete files
- Demo effect of dropping table

```sql
%fs ls '/Volumes/gizmbobox/landing/landing/external_data/payments/'
```

```sql
-- Create External Data
CREATE TABLE IF NOT EXISTS gizmbobox.bronze.payments AS
SELECT
    CAST(_c0 AS INTEGER) AS payment_id,
    CAST(_c1 AS INTEGER) AS order_id,
    CAST(_c2 AS TIMESTAMP) AS payment_timestamp,
    CAST(_c3 AS INTEGER) AS payment_status,
    _c4 AS payment_METHOD
FROM read_files(
    '/Volumes/gizmbobox/landing/landing/external_data/payments/',
    format => 'csv',
    header => false
);
```
